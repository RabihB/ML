```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('dataset/dataset1.csv')
df.head()
```

```python
x = df[['x', 'y']].values
y = df['z'] == "a"
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y);
```

### StandardScaler

```python
from sklearn.preprocessing import StandardScaler     #scales at (m-mean)/stdm
scaledx = StandardScaler().fit_transform(x)

plt.scatter(scaledx[:, 0], scaledx[:, 1], c = y);    
```

### QuantileTransformer

```python
from sklearn.preprocessing import QuantileTransformer     #Scales using quantiles   
scaledx = QuantileTransformer(n_quantiles=100).fit_transform(x)

plt.scatter(scaledx[:, 0], scaledx[:, 1], c = y);
```

### OneHotEncoder

```python
arr = np.array(['low', 'low', 'high', 'medium']).reshape(-1, 1)
arr
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse=False, handle_unknown ='ignore')
enc.fit_transform(arr)
```
